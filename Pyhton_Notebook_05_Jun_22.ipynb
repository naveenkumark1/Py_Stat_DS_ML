```python
## Simple calculator using python 

## take input of 2 numbers from user 
## take input of operation from user (add, subt, mult, divi)
## give the output


def add(x,y):
    return x+y
def subt(x,y):
    return x-y
def multi(x,y):
    return x*y
def divi(x,y):
    return x/y

print(" Select the operation")
print("1- Add, 2- Sub, 3-Mult, 4- Div")

while True:
    inp = input(" Enter choice 1,2,3,4")
    
    if inp in ('1','2','3','4'):
        n1 = float(input("Enter first number"))
        n2 = float(input("Enter first number"))
        
        if inp =='1':
            print(add(n1,n2))
        elif inp =='2':
            print(subt(n1,n2))
        elif inp =='3':
            print(multi(n1,n2))
        elif inp =='4':
            print(divi(n1,n2))
            
        next = input("Do you want to do another calclutation ?? Y/N")
        
        if next == 'N':
            break
    else:
        print("Not a valid input")
```
